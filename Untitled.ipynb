```python
##Remove missing value of actual_delivery_time ,bcz data is huge and missing value is so small 
df=df[~df['actual_delivery_time'].isna()]
```

```python
##Convert string into DateTime 
df['created_at'] = pd.to_datetime(df['created_at'])
df['actual_delivery_time']=pd.to_datetime(df['actual_delivery_time'])
```

```python
df[['created_at','actual_delivery_time']].info()
```

```python
##Extracting second from columns for target 
date_col = ['created_at','actual_delivery_time']

df['created_sec']=df['created_at'].dt.second
df['actual_delivery_time_sec']=df['actual_delivery_time'].dt.second
```

```python
df.drop([['created_at','actual_delivery_time']],axis=1,inplace=True)
```

```python
df.sample()
```

```python
##Target varible 
df['Target'] = df['actual_delivery_time_sec'] - df['created_sec']
df['Target'] = df['Target'].abs()
```

```python
##New feature might be useful 

df['Estimtaed_non_prep_time'] = df['estimated_order_place_duration'] + df['estimated_store_to_consumer_driving_duration']
```
